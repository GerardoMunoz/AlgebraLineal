# Transpuesta
## Definición
La __transpuesta__ de de una matriz $A$ de tamaño $ m \times n$ 
$$ A=\begin{bmatrix}
a_{00} & a_{01} & \cdots & a_{0(n-2)} & a_{0(n-1)} \\
\vdots & \vdots & \ddots & \vdots & \vdots \\
a_{(m-1)0} & a_{(m-1)1} & \cdots & a_{(m-1)(n-2)} & a_{(m-1)(n-1)} \\
\end{bmatrix}$$

Se denota $A^T$ y es la matriz de tamaño $n \times m$  
$$ A^T=\begin{bmatrix}
a_{00} &  \cdots & a_{(m-1)0} \\
a_{01} &  \cdots & a_{(m-1)1} \\
\vdots
& \ddots & \vdots\\
a_{0(n-2)} &  \cdots & a_{(m-1)(n-2)} \\
a_{0(n-1)} &  \cdots & a_{(m-1)(n-1)} \\
\end{bmatrix}$$
cuyo elemento $ij$ es $a_{ji}$. Entonces la primera columna (renglón) de $A^T$ es el primer renglón (columna) de $A$ y así sucesivamente.

Por ejemplo:
$$\begin{bmatrix}
1 & 5 & 3 & 8\\
2 & 7 & 11 & 6
\end{bmatrix}^T
=\begin{bmatrix}
1 & 2 \\
5 & 7 \\
3 & 11 \\
8 & 6 
\end{bmatrix}$$

## Propiedades

Sean $A$ y $B$ matrices de tamaños adecuados, $c$ un escalar y $u$ y $v$ vectores de tamaños adecuados.

* Transpuesta del producto: $(AB)^{T}=B^{T}A^{T}$
* Transpuesta de la suma: $(A+B)^{T}=A^{T} + B^{T}$
* Transpuesta de la transpuesta: $(A^{T})^{T}=A$
* Transpuesta de escalar por matriz $(cA)^{T}=c(A)^{T}$
* Inversa y transpuesta conmutan: Si $A$ es invertible entonces $A^T$ también lo es y $(A^T)^{-1}=(A^{-1})^T$
* Transpuesta y producto punto: $(Au) \cdot v = u \cdot (A^T v)$

Usando la anterior notación tenemos las siguientes propiedades:
* $\rho(T_A)=\rho(T_{A^T})$
* $\rho(T_{A^T}) + \nu(T_A) = n$
* $\rho(T_A) + \nu(T_{A^T}) = m$
* $\text{Im}(T_{A^T}) \perp \text{Nu}(T_A) $
* $\text{Im}(T_A) \perp \text{Nu}(T_{A^T}) $

Sean $W_0$ y $W_1$ dos subespacios del espacio vectorial $V$, entonces $W_0$ y $W_1$ son __subespacios ortogonales__   ($W_0 \perp W_1$) si para todo $w_0 \in W_0$ y para todo $ w_1 \in W_1 $ se tiene que $w_0 \cdot w_1=0$.

Sea $W$ un subespacio de un espacio vectorial $V$. El __complemento ortogonal__ de $W$ es el conjuto más grande ortogonal a $W$, el cual se puede definir de la siguiente forma.
$$W^{\perp}=\{v \in V \mid v \cdot w =0 \text{ para todo }w \in W\}$$
El complemento ortogonal cumple las siguientes propiedades:

* $ W \perp W^{\perp} $ 
* $\text{Dim}(W) + \text{Dim}(W^{\perp}) = \text{Dim}(V) $
* $ W \cap W^{\perp}={\vec{0}} $

## Matriz y conjunto ortogonal y ortonormal

Sea $A$ una matriz cuadrada con columnas $v_0, v_1, \ldots , v_{(n-1)}$. 
Para que la transpuesta  
$$A^T=\begin{bmatrix}v_0 \\ v_1 \\ \vdots \\ v_{(n-1)} \end{bmatrix}$$ 
sea la inversa 
$$A^{-1}=\begin{bmatrix}v_0 & v_1 & \cdots & v_{(n-1)} \end{bmatrix}^{-1}$$ 

se debe cumplir que

\begin{align}
A^T A  &= I\\
\begin{bmatrix}v_0 \\ v_1 \\ \vdots \\ v_(n-1) \end{bmatrix}\begin{bmatrix}v_0 & v_1 & \cdots & v_(n-1) \end{bmatrix}
&=\begin{bmatrix}1 & 0 & \cdots & 0 \\ 0 & 1 & \cdots & 0 \\ \vdots \\ 0 & 0 & \cdots & 1 \end{bmatrix}\\ 
\begin{bmatrix}v_0 \cdot v_0 & v_0 \cdot v_1 & \cdots & v_0 \cdot v_{(n-1)} \\ v_1 \cdot v_0 & v_1 \cdot v_1 & \cdots & v_1 \cdot v_{(n-1)} \\ \vdots \\ v_{(n-1)} \cdot v_0 & v_{(n-1)} \cdot v_1 & \cdots & v_{(n-1)} \cdot v_{(n-1)} \end{bmatrix}
&=\begin{bmatrix}1 & 0 & \cdots & 0 \\ 0 & 1 & \cdots & 0 \\ \vdots \\ 0 & 0 & \cdots & 1 \end{bmatrix}\\ 
\end{align}

Por lo tanto para $i,j \in \{0, 1,  \ldots , n-1 \}$
$$v_i \cdot v_j = 
\left\{ 
\begin{aligned}
&1 \text{ si } i=j\\
&0 \text{ si } i \neq j
\end{aligned}
\right.
$$

Lo que significa que 
* $|v_i|=1$, $v_i$ es un vector unitario 
* $v_i \perp v_j $ si $i \neq j$ 

El conjunto $\{v_0, v_1, \cdots,  v_{(n-1)}\}$ con estas propiedades se conoce como __conjunto ortonormal__ y la respectiva matriz cuadrada $A=\begin{bmatrix}v_0 & v_1 & \cdots & v_{(n-1)} \end{bmatrix}$ se debería llamar matriz ortonormal, pero históricamente se conoce como __matriz ortogonal__.

Un conjunto  $\{v_0, v_1, \cdots,  v_{(k-1)}\}$  de de vectores de $\mathbb{R}^n$ sólo con la propiedad $v_i \perp v_j $ si $i \neq j$ se llama __conjunto ortogonal__. 

## Propiedades de los conjuntos ortogonales

Sea $S=\{v_0, v_1,  \cdots,  v_{(k-1)}\}$ un conjunto ortogonal de vectores distintos de cero y sea 
$$u=c_0 v_0 + c_1 v_1 + \cdots + c_{(n-1)} v_{(n-1)} $$ 
en donde $c_0, c_1, \ldots, c_{(n-1)} \in \mathbb{R} $
entonces: 
* $c_i=\frac{u \cdot v_i}{v_i \cdot v_i} \text{ con } i \in \{0,1, \ldots , (n-1) \}$
* $S$ es un conjunto L.I.
* $R=\{w_0, w_1, \cdots,  w_{(n-1)}\}$ con $w_i=\frac{1}{|v_i|}v_i$ es un conjunto ortonormal. 
* Si $\{w_0, w_1, \cdots,  w_{(n-1)}\}$ es un conjunto ortonormal entonces  $c_i=u \cdot w_i  \text{ con } i \in \{0,1, \ldots , (n-1) \}$

### Definición
Sea $u$ un vector de $\mathbb{R}^n$ y sea $V$ un subespacio de $\mathbb{R}^n$ con una base ortogonal  
 $B=\{v_0, v_1, \ldots v_{(k-1)} \}$. La __proyección ortogonal__ de $u$ sobre $V$ es el vector
 $$ u_{pr} =\frac{u \cdot v_0}{v_0 \cdot v_0} v_0 + \frac{u \cdot v_1}{v_1 \cdot v_1} v_1 + \cdots + \frac{u \cdot v_{(k-1)}}{v_{(k-1)} \cdot v_{(k-1)}} v_{(k-1)}$$ 

## Gram Smidt

Teorema: [Nakos,Teorema 8.2.16]

Si $B=\{v_0, v_1, \ldots v_{(k-1)} \}$ es una base de $V$ entonces $B'=\{u_0, u_1, \ldots u_{(k-1)} \}$ es una base ortogonal, donde 

\begin{align}
u_0 &= v_0\\
u_1 &= v_1 - \frac{v_1 \cdot u_0}{u_0 \cdot u_0}u_0\\
u_2 &= v_2 - \frac{v_2 \cdot u_0}{u_0 \cdot u_0}u_0 - \frac{v_2 \cdot u_1}{u_1 \cdot u_1}u_1\\
&\vdots\\
u_{(k-1)} &= v_{(k-1)} - \frac{v_{(k-1)} \cdot u_0}{u_0 \cdot u_0}u_0 - \frac{v_{(k-1)} \cdot u_1}{u_1 \cdot u_1}u_1 - \cdots - \frac{v_{(k-1)} \cdot u_{(k-2)}}{u_{(k-2)} \cdot u_{(k-2)}}u_{(k-2)}
\end{align}
y una base ortonormal $B''$ se obtiene normalizando a $B'$.
$$B''=\left\{ \frac{1}{|u_0|}u_0, \frac{1}{|u_1|}u_1, \ldots , \frac{1}{|u_{(k-1)}|}u_{(k-1)} \right\}$$
