# √Årea del paralelogramo en 2D

```python
import matplotlib.pyplot as plt
from sympy import *
from ubogsla18p import *
x,y=symbols('x y')
o2=Matrix([0,0])
imprimir('o2=',o2)
```

Dados los vectores

```python
u=Matrix([0.6,0.3])
imprimir('u=',u)
```

```python
v=Matrix([0.1,0.2])
imprimir('v=',v)
```

Podemos dibujar el paralelogramo formado por los dos vectores

```python
ax=ejes(0,0,1,0.6)
flecha(ax,v,nombre='$v$') 
flecha(ax,u,nombre='$u$') 
flecha(ax,v,u,nombre='$u+v$') 
flecha(ax,u,v) 
angulo(ax,u,v,0.3,nombre=r'$\alpha$')
plt.show()
```

El objetivo es encontrar el √°rea de ese paralelogramo en 2D.
$$\text{Area}=bh$$
Donde 
* $b$ es la base, $b=|u|$.
* $h$ es la altura del paralelogramo $h=|v| \text{sin}(\alpha)$$

```python
ax=ejes(0,0,1,0.6)
flecha(ax,v,nombre='$v$') 
flecha(ax,u,nombre='$u$') 
flecha(ax,v,u,nombre='$u+v$') 
flecha(ax,u,v) 
uvp=(u.dot(v)/u.dot(u))*u
uvo=v-uvp
flecha(ax,uvo,uvp,color='r')
angulo(ax,u,v,0.2,nombre=r'$\alpha$')
plt.show()
```

$$\text{sin}(\alpha)=\text{cos}(\alpha-\frac{\pi}{2})$$
Al colocar todo junto 
$$\text{Area}=|u|\ |v|\ \text{cos}(\alpha-\frac{\pi}{2})$$
La ecuaci√≥n anterior recuerda el producto punto. Sin embargo, el √°ngulo est√° rotado -90¬∞. Para solucionar esto podemos rotar -90¬∞ el vector $v$.

```python
#Matriz de rotaci√≥n de -pi/2
alpha=-pi/2
R=Matrix([[cos(alpha), -sin(alpha)],[sin(alpha), cos(alpha)]])
xy=Matrix([x,y])
imprimir(R,xy,'=',R*xy)
```

Para rotar un vector -90¬∞ basta con intercambiar los renglones y luego cambiar el signo al segundo rengl√≥n. 

```python
w=R*v
imprimir('w=',R,v,'=',w)
```

```python
ax=ejes(0,-0.3,1,0.6)
flecha(ax,v,nombre='$v$') 
flecha(ax,u,nombre='$u$') 
flecha(ax,v,u,nombre='$u+v$') 
flecha(ax,u,v) 
flecha(ax,w,nombre='$w$') 
angulo(ax,u,v,0.2,nombre=r'$\alpha$')
angulo(ax,w,u,0.25,nombre=r'$\beta=\alpha-90¬∞$')
plt.show()
```

```python
ax = plt.axes((0,0,1.5,1.3))
ax.set_xlim(-1,1)
ax.set_ylim(-0.3,1)
x1=float(v[0])
flecha(ax,v,nombre='$v$') 
flecha(ax,u,nombre='$u$') 
flecha(ax,v,u,nombre='$u+v$') 
flecha(ax,u,v) 
flecha(ax,w,nombre='$w$') 
angulo(ax,u,v,0.5,nombre=r'$\alpha$')
angulo(ax,w,u,0.4,nombre=r'$\beta=\alpha-90¬∞$')
plt.show()
```

```python
ax=ejes(0,-0.3,1,0.6)
flecha(ax,v,nombre='$v$') 
flecha(ax,u,nombre='$u$') 
flecha(ax,v,u,nombre='$u+v$') 
flecha(ax,u,v) 
flecha(ax,w,nombre='$w$') 
angulo(ax,u,v,0.2,nombre=r'$\alpha$')
angulo(ax,w,u,0.25,nombre=r'$\beta=\alpha-90¬∞$')
proy_v_u=(u.dot(v)/u.dot(u))*u
ort_v_u=v-proy_v_u
flecha(ax,ort_v_u,proy_v_u,color='r')
proy_w_u=(u.dot(w)/u.dot(u))*u
flecha(ax,proy_w_u,color='r')
plt.show()
```

La altura del paralelogramo es igual a la proyecci√≥n del vector $w$ sobre el vector $u$
$$|v|\sin(\alpha)=|w|\cos(\beta)$$

Esto permite calcular el √°rea del paralelogramo como un producto punto. Supongamos que $u=\begin{bmatrix}a\\c\end{bmatrix}$ y si $v=\begin{bmatrix}b\\d\end{bmatrix}$ entonces $w=\begin{bmatrix}0 & -1\\1 & 0\end{bmatrix}\begin{bmatrix}b\\d\end{bmatrix}=\begin{bmatrix}d\\-b\end{bmatrix}$. Por lo tanto, el √°rea del paralelogramo \begin{align}
\text{√Årea}
&=|u|\ |v|\sin(\alpha)\\
&=|u|\ |w|\cos(\beta)\\
&=u \cdot w\\
&=\begin{bmatrix}a\\c\end{bmatrix} \cdot \begin{bmatrix}d\\-b\end{bmatrix}\\
&=ad-cb
\end{align}
El √°rea formada por los vectores $u$ y $v$ se denota como el determinante de la matriz de $2 \times 2$ $$\begin{vmatrix}a & b\\c & d\end{vmatrix}=ad-cb$$
