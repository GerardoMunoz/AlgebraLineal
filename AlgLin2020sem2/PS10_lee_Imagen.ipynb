```
import matplotlib.pyplot as plt
import cv2
import numpy as np
import requests

# perform request
response =  requests.get('https://docs.google.com/uc?export=download&id=1Ne8aWW5QZ_QnpHyC1pffaqi3ON9kHE6z').content
# convert to array of ints
nparr = np.frombuffer(response, np.uint8)
# convert to image array
img = cv2.imdecode(nparr,0)
#lee el tamaño de la imagen
h,w=img.shape
# coloca marco
img = np.pad(img,max(h//2,w//2))
#lee en nuevo tamaño de la imagen
h,w=img.shape
# showimage
plt.imshow(img, cmap = 'gray')#, interpolation = 'bicubic')
plt.show()
```

```
#Hace la imagen el doble de ancha
I=np.float32([[2,0,0],
              [0,1,0],
              [0,0,1]])
print(I[:2,:])
dst = cv2.warpAffine(img,I[:2,:],(w,h))
plt.imshow(dst, cmap = 'gray')
plt.show()
```
