# Norma

## Definición:

Sea $V$ un espacio vectorial real. Un __norma__ es una función $ | \cdot | :V \rightarrow \mathbb{R}$ que satisface tres axiomas. Sean $u,v \in V$ y $a \in \mathbb{R}$:

1. $| u | \geq 0$, con igualdad sólo si $u=\vec{0}$
1. $| au | = \text{abs}(a)| u |$
1. $| u + v | \leq |u| + |v|$

El último renglón de la tabla generaliza los tres renglones anteriores. Cada una de estas normas define un subespacio vectorial, en donde la norma converge para cada elemento del subespacio.


$$\mathcal{L}^p=\left\{u:\mathbb{R} \rightarrow \mathbb{R} \mid \sqrt[p]{\int_{-\infty}^{\infty} |u(t)|^p \ dt } < \infty\right\}$$

$$
{l}^p=\left\{u:\mathbb{Z} \rightarrow \mathbb{R} \mid \sqrt[p]{\sum_{k=-\infty}^{\infty} |u[k]|^p} < \infty\right\}$$


# Producto interno

## Definición:

Sea $V$ un espacio vectorial real. Un __producto interno__ es una función $ \langle \cdot , \cdot \rangle :V \times V \rightarrow \mathbb{R}$ que satisface tres axiomas. Sean $u,v,w \in V$ y $a,b \in \mathbb{R}$:

1. $\langle u , v \rangle = \langle v , u \rangle^*$
2. $\langle au + bv , w \rangle = a\langle u , w  \rangle + b\langle v , w \rangle$
3. $\langle u , u \rangle \geq 0$, con igualdad sólo si $u=\vec{0}$

$^*$ En el caso de los números complejos se utiliza el complejo conjugado

## Ejemplo:
A continuación se listan algunos espacios vectoriales con una propuesta de producto interno.

| Espacio vectorial | ---------------------------------------- Producto interno ---------------------------------------- |
|-|-|
| $$V= \mathbb{R}^2$$ | $$\langle u , v \rangle = \left\langle \begin{bmatrix} u_0 \\ u_1 \end{bmatrix} , \begin{bmatrix} v_0 \\ v_1 \end{bmatrix} \right\rangle= u_0 v_0 +u_1 v_1$$ |
| $$V= \mathbb{R}^3$$ | $$\langle u , v \rangle = \left\langle \begin{bmatrix} u_0 \\ u_1 \\ u_2 \end{bmatrix} , \begin{bmatrix} v_0 \\ v_1 \\ v_2 \end{bmatrix} \right\rangle= u_0 v_0 +u_1 v_1+u_2 v_2$$ |
| $$V= \mathbb{R}^n$$ | $$\langle u , v \rangle = \left\langle \begin{bmatrix} u_0 \\ u_1 \\ \vdots \\ u_{(n-1)} \end{bmatrix} , \begin{bmatrix} v_0 \\ v_1 \\ \vdots \\ v_{(n-1)} \end{bmatrix} \right\rangle= \sum_{k=0}^{(n-1)} u_k v_k 
$$|
| $$V= \mathbb{R}^\mathbb{Z}$$ | $$\langle u , v \rangle =  \sum_{k=-\infty}^{\infty} u[k] v[k] 
$$|
| $$V= \mathbb{R}^\mathbb{R}$$ | $$\langle u , v \rangle =  \int_{-\infty}^{\infty} u(t) v(t) dt 
$$|

En los dos últimos casos es preocupante que la sumatoria o la integral no converjan, por este motivo es conveniente colocarle condiciones adicionales al espacio vectorial para que converja el respectivo producto interno.

| Nombre | ----- Dimensión $\mathbb{R}$ (continuo) ----- | ----- Dimensión $\mathbb{z}$ (discreto) ----- |
|-|-|-|
| Sin condición | $$V= \mathbb{R}^\mathbb{R}$$ | $$V= \mathbb{R}^\mathbb{Z}$$ |
| Absolutamente sumables | ($V= \mathbb{R}^\mathbb{R}$) y ($\int_{-\infty}^{\infty} |u(t)| \ dt \leq \infty$) | ($V= \mathbb{R}^\mathbb{Z}$) y ($\sum_{k=-\infty}^{\infty} |u[k]| \leq \infty$)| 
| De energía | ($V= \mathbb{R}^\mathbb{R}$) y ($\int_{-\infty}^{\infty} |u(t)|^2 \ dt \leq \infty$) | ($V= \mathbb{R}^\mathbb{Z}$) y ($\sum_{k=-\infty}^{\infty} |u[k]|^2 \leq \infty$)| 
| Acotada | ($V= \mathbb{R}^\mathbb{R}$) y ($\underset{t \in \mathbb{R}}{\text{max}} (u(t)) \leq \infty$) | ($V= \mathbb{R}^\mathbb{Z}$) y ($\underset{k \in \mathbb{Z}}{\text{max}} (u[k]) \leq \infty$)| 

Un producto interno en $V= \mathbb{R}^2$ es el producto punto, definido de la siguiete forma para cualquier $u,v \in \mathbb{R}$.
$$\langle u , v \rangle = \langle \begin{bmatrix} u_0 \\ u_1 \end{bmatrix} , \begin{bmatrix} v_0 \\ v_1 \end{bmatrix} \rangle= u_0 v_0 +u_1 v_1$$
