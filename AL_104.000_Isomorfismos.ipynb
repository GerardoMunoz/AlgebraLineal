# Isomorfismos

## Imagen y Nucleo

## Definición:
Sea $T:V \rightarrow W$ una transformación lineal.

El __núcleo__ de $T$ es un subconjunto de $V$ formado por los vectores que son transformados en el cero. En algunos también se llama __espacio nulo__.

$$\text{Nu}(T)=\{v \in V \mid T(v)=\vec{0}\}$$

La __imagen__ de $T$ es un subconjunto de $W$ formado por los vectores que son transformados por algún vector de $V$. 


$$\text{Im}(T)=\{T(v) \mid v \in V \}$$

En el caso de que $W=R^n$ y $V=R^m$, entonces existe una matriz $A$ de tamaño $m \times n$ tal que $T(v)= Av$. En este caso la imagen se puede llamar el __espacio columna__ de $A$ o  __espacio generado__ por las columnas de $A$. 

$$\text{Col}(A)=\text{Gen(A)}=\text{Im}(T_A)$$


## Teorema:
* La imagen de $T_A$ es un subespacio vectorial de $W$.
* El núcleo de $T_A$ es un subespacio vectorial de $V$.



## Transformaciones lineales inyectivas  y sobreyectivas
## Definición:
Sea $T:V \rightarrow W$ una transformación.
* $T$ es __inyectiva__ _sii_ $v_1 \neq v_2$ entonces  $T(v_1) \neq T(v_1)$ para todo $v_1, v_2 \in V$. 
* $T$ es __sobreyectiva__ _sii_ para todo $w \in W$ existe un  $v \in V$ tal que   $T(v)=w$.
* $T$ es __biyectiva__  _sii_ $T$ es inyectiva y sobreyectiva.

Si $T:V \rightarrow W$ es una T.L. biyectiva entonces se dice que $T$ es un __isomorfismo__ entre espacios vectoriales y que los espacios vectoriales $V$ y $W$ son __isomorfos__.

## Teorema:
* La composición de transformaciones inyectivas es inyectiva.
* La composición de transformaciones sobreyectivas es sobreyectiva.
* La composición de transformaciones biyectivas es biyectiva.

## Teorema:
* $T_A$ es sobreyectiva _sii_ Im$(T_A)=W$ 
* $T_A$ es inyectiva _sii_ Nu$(T_A)=\{\vec{0}\}$


# Definición:
Sea $S:R^n \rightarrow V$ un isomorfismo, entonces:
* $S^{-1}(v)$ se llama las __coordenadas__ de $v$ en la base $B$ y se denota
$[v]_B=S_B^{-1}(v)$
* La __dimensión__ del espacio vectorial $V$ es el número de elementos de la base, es decir $n$. dim($V$)=$n$.

<a id='dim0'> </a>
Un conjunto que contiene solamente un vector cero, $\{\vec{0}\}$, es el espacio vectorial más pequeño. La dimensión de este espacio vectorial es el número $0$. $$\text{Dim}(\{\vec{0}\})=0$$  
Por lo tanto, la base de $\{\vec{0}\}$ no puede no puede tener vectores, entonces es una sucesión vacía $B=[]$.


Si un espacio $V$ no es isomorfo a ningún espacio $R^n$ entonces se dice que $V$ no tiene dimensión finita. Por ejemplo, no tienen dimensión finita los espacios vectoriales:
* de los polinomios de cualquier grado $P$,
* de las funciones de valor real $F(R)$.

## Definición:
* El __rango__ de $T_A$ (o el rango de $A$) es la dimensión de su imagen. $$\rho(T_A)=\rho(A)=\text{Dim}(\text{Im}(T_A)).$$
* La __nulidad__ de $T_A$ (o la dimensión de $A$) es la dimensión de su núcleo. $$\nu(T_A)=\nu(A)=\text{Dim}(\text{Nu}(T_A)).$$



## Definición:

Sea $V$ un espacio vectorial y  $v_1, v_2 \ldots v_n \in V$.

*  $[v_1, v_2 \ldots v_n] $ es __Linealmente Independiente__ (L.I.) _sii_ existen coeficientes únicos $c_1, c_2, \ldots c_n$ tales que 
$$ c_1 v_1 + c_2 v_2 + \cdots + c_n v_n = \vec{0}$$

* $[v_1, v_2 \ldots v_n]$ __genera__ a $V$, (denotado gen($[v_1, v_2 \ldots v_n])=V$) _sii_ para cada vector
$u \in V$ se tiene que existen coeficientes $c_1, c_2, \ldots c_n$ tales que 
$$ c_1 v_1 + c_2 v_2 + \cdots + c_n v_n = u$$

* $[v_1, v_2 \ldots v_n] $ es __base__ de $V$ _sii_ es L.I. y genera a $V$. Es decir, 
si para cada vector $u \in V$ se tiene que existen coeficientes únicos $c_1, c_2, \ldots c_n$ tales que 
$$ c_1 v_1 + c_2 v_2 + \cdots + c_n v_n = u$$



## Teorema:
* $T:R^n \rightarrow V$ es una T.L. inyectiva _sii_ $T^*(I)$ es 
Linealmente Independiente.
* $T:R^n \rightarrow V$ es una T.L. sobreyectiva _sii_ $T^*(I)$ genera a $V$, Gen$(T^*(I))=V$.
* $T:R^n \rightarrow V$ es un isomorfismo _sii_ $T^*(I)$ es una base de $V$.
